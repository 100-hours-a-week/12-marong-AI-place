```python
import ast
import numpy as np
import re
import emoji
from sentence_transformers import SentenceTransformer

# 모델 로드
model = SentenceTransformer('snunlp/KR-SBERT-V40K-klueNLI-augSTS')

# 불용어 리스트
stopwords = set(['것', '정말', '좀', '너무', '매우', '입니다', '하는', '하고'])

# 단어 반복 줄이기
def reduce_repeated_words(text):
    return re.sub(r'(\b\w+)\1{1,}', r'\1', text)

# 한국어 텍스트 클린 함수
def clean_korean_text(text):
    if not text:
        return ''
    text = emoji.replace_emoji(text, replace='')
    text = re.sub(r'[^가-힣\s]', '', text)
    text = reduce_repeated_words(text)
    words = text.split()
    words = [word for word in words if word not in stopwords]
    return ' '.join(words)

# 리뷰 평균 임베딩 함수
def embed_reviews_avg(review_str):
    try:
        review_dict = ast.literal_eval(review_str)
    except:
        return None

    review_texts = []
    for review_info in review_dict.values():
        content = review_info.get('리뷰 내용', '')
        if content and content != '리뷰 없음':
            cleaned_content = clean_korean_text(content.strip())  # ✅ 여기 추가
            if cleaned_content:  # 전처리 후에도 내용이 남아있을 때만
                print(cleaned_content)
                review_texts.append(cleaned_content)

    if not review_texts:
        return None

    embeddings = []
    for text in review_texts:
        prompt = f"이 리뷰가 묘사하는 분위기를 임베딩하세요: {text}"
        emb = model.encode(prompt)
        embeddings.append(emb)

    avg_embedding = np.mean(embeddings, axis=0)
    return avg_embedding.tolist()  # ✅ 리스트로 변환해서 저장

# 적용
df['리뷰_평균임베딩'] = df['리뷰'].apply(embed_reviews_avg)
```

```python
from sentence_transformers import SentenceTransformer
import ast

# SBERT 모델 로드
model = SentenceTransformer('snunlp/KR-SBERT-V40K-klueNLI-augSTS')

# 메뉴 임베딩 함수
def embed_menu(row):
    category = row['대표_카테고리']  # 한식, 일식 등

    # 메인_메뉴 우선
    try:
        main_menu = ast.literal_eval(row['메인_메뉴_형태']) if not pd.isna(row['메인_메뉴_형태']) else []
    except:
        main_menu = []

    # 메인_메뉴 없으면 메뉴 전체를 사용
    if not main_menu:
        try:
            full_menu = ast.literal_eval(row['메뉴']) if not pd.isna(row['메뉴']) else []
        except:
            full_menu = []
        menu_list = full_menu
    else:
        menu_list = main_menu

    if not category or not menu_list:
        return None

    # 자연어 프롬프트 생성
    prompt = f"이 식당은 {category} 전문점입니다. 주요 메뉴로는 {', '.join(menu_list)}가 있습니다. 이 식당의 음식 스타일을 임베딩하세요."

    # SBERT 임베딩
    return model.encode(prompt).tolist()

# 적용
df['메뉴_임베딩'] = df.apply(embed_menu, axis=1)
```

```python
import pandas as pd
import numpy as np
import ast
import chromadb
from chromadb.config import Settings
import torch.nn.functional as F
import torch

# ✅ 정규화 함수 추가
def normalize_vector(vec):
    return F.normalize(torch.tensor(vec, dtype=torch.float32), dim=0).tolist()


# ✅ ChromaDB 로컬 연결
chroma_client = chromadb.PersistentClient(
    path="./chroma_db",  # 로컬 폴더
    settings=Settings(anonymized_telemetry=False)
)

# ✅ 두 개 컬렉션 생성
review_collection = chroma_client.get_or_create_collection(name="review_collection")
menu_collection = chroma_client.get_or_create_collection(name="menu_collection")

# ✅ 데이터 불러오기
df = pd.read_csv('/Users/kimss/Documents/marong/임베딩_완료.csv')

# ✅ 벡터 파싱 함수
def parse_embedding(emb_str):
    try:
        emb = ast.literal_eval(emb_str)
        return np.array(emb)
    except:
        return None

# ✅ 컬럼 파싱
df['리뷰_평균임베딩'] = df['리뷰_평균임베딩'].apply(parse_embedding)
df['메뉴_임베딩'] = df['메뉴_임베딩'].apply(parse_embedding)

# ✅ 유효 데이터 필터링
valid_documents = []
valid_review_embs = []
valid_menu_embs = []
valid_metadatas = []
valid_ids = []

for idx, row in df.iterrows():
    review_emb = row['리뷰_평균임베딩']
    menu_emb = row['메뉴_임베딩']

    if review_emb is not None and menu_emb is not None:
        valid_documents.append(row['상호명'])
        valid_review_embs.append(normalize_vector(review_emb))
        valid_menu_embs.append(normalize_vector(menu_emb))

        metadata = {
            "상호명": row.get('상호명', ''),
            "대표카테고리": row.get('대표_카테고리', ''),
            "메인_메뉴": row.get('메인_메뉴_형태', ''),
            "주소": row.get('주소', ''),
            "평균별점": row.get('평균별점', ''),
            "영업시간": row.get('영업시간_요약', ''),
            "위도": row.get('위도', ''),
            "경도": row.get('경도', ''),
            "링크": row.get('링크', '')
        }
        valid_metadatas.append(metadata)
        valid_ids.append(f"store_{idx}")

# ✅ 배치 업로드 함수
def add_to_chroma_in_batches(collection, documents, embeddings, metadatas, ids, batch_size=500):
    for i in range(0, len(documents), batch_size):
        batch_docs = documents[i:i+batch_size]
        batch_embs = embeddings[i:i+batch_size]
        batch_meta = metadatas[i:i+batch_size]
        batch_ids = ids[i:i+batch_size]

        collection.add(
            documents=batch_docs,
            embeddings=batch_embs,
            metadatas=batch_meta,
            ids=batch_ids
        )
        print(f"✅ Added batch {i // batch_size + 1}")

# ✅ 업로드 실행
add_to_chroma_in_batches(review_collection, valid_documents, valid_review_embs, valid_metadatas, valid_ids)
add_to_chroma_in_batches(menu_collection, valid_documents, valid_menu_embs, valid_metadatas, valid_ids)

# ✅ 문서 수 확인
print("리뷰 컬렉션 문서 수:", review_collection.count())
print("메뉴 컬렉션 문서 수:", menu_collection.count())
```

```python
import pandas as pd
import numpy as np
import chromadb
from chromadb.config import Settings

# ✅ ChromaDB 연결
client = chromadb.PersistentClient(path="/Users/kimss/Documents/marong/chroma_db")  # 너가 지정한 경로

# ✅ 컬렉션 로딩
menu_col = client.get_collection("menu_collection")
review_col = client.get_collection("review_collection")

# ✅ 모든 document 가져오기
menu_docs = menu_col.get(include=["metadatas"])
review_docs = review_col.get(include=["metadatas"])

# ✅ 각 document의 metadata 추출
menu_df = pd.DataFrame(menu_docs["metadatas"])
review_df = pd.DataFrame(review_docs["metadatas"])

# ✅ place_id 기준 merge
merged_df = pd.merge(menu_df, review_df, on="링크", suffixes=("_menu", "_review"))

# ✅ 필요한 컬럼 정제
df = merged_df[["상호명_menu", "평균별점_menu", "링크", "경도_menu", "위도_menu"]].copy()

# ✅ 거리 역수 처리 (가까울수록 좋은 특성으로 만들기)
df["distance"] = 1 / df["distance"]

# ✅ 정규화 및 엔트로피 가중치 계산
def calculate_entropy_weights(df):
    norm_df = df / df.sum()
    k = 1 / np.log(len(df))
    entropy = -k * (norm_df * np.log(norm_df + 1e-12)).sum()
    diversity = 1 - entropy
    weights = diversity / diversity.sum()
    return weights

# ✅ 엔트로피 기반 가중치 계산
feature_df = df[["menu_similarity", "review_similarity", "rating", "distance"]]
weights = calculate_entropy_weights(feature_df)

# ✅ 출력
print("🚀 엔트로피 기반 가중치 결과:")
for col, w in weights.items():
    print(f"{col} 가중치: {w:.4f}")
```
